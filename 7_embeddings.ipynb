```python
import pandas as pd
import nltk
import string
import matplotlib.pyplot as plt

from nltk.corpus import stopwords
from nltk import word_tokenize
from gensim.models import Word2Vec as w2v
from sklearn.decomposition import PCA

import numpy as np
import json
```

```python
def read_json(filename):    
    with open(filename, 'r') as openfile:
        return json.load(openfile)

def write_json(to_json, filename):
    json_object = json.dumps(to_json)

    with open(filename, "w") as outfile:
        outfile.write(json_object)
```

```python
lemmatized_filtered_tokens = read_json("2_lemmatized_filtered_black_clover.json")
```

```python
lemmatized_filtered_tokens
```

```python
w = w2v(
    [lemmatized_filtered_tokens],
    min_count=1,  
    sg = 1,       
    window=7      
)
```

```python
print(w.wv.most_similar('magic'))
```

```python
word2vecfied = {str(n):[float(x) for x in w.wv.get_vector(str(n))] for n in w.wv.key_to_index}
```

```python
word2vecfied['Yuno']
```

```python
w.wv.most_similar(np.array(word2vecfied['magic']))
```

```python
write_json(word2vecfied, "7_word2vecfied_black_clover.json")
```
