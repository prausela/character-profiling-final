### Indexing

```python
tokens      = jsu.read_json("9_tokens_black_clover.json")
characters  = jsu.read_json("3_characters_black_clover.json")
```

```python
(index2token, token2index)         = encu.index(tokens)
(index2character, character2index) = encu.index(characters)
```

```python
jsu.write_json(index2token, "10_index2token_black_clover.json")
jsu.write_json(token2index, "10_token2index_black_clover.json")
```

```python
jsu.write_json(index2character, "10_index2character_black_clover.json")
jsu.write_json(character2index, "10_character2index_black_clover.json")
```

### Encoding sentences

```python
tokenized_labeled_sentences = jsu.read_json("9_tokenized_sentences_black_clover.json")
tokenized_labeled_sentences
```

```python
hot_encoded_tokenized_labeled_sentences = [
    {
        "subjects"  : encu.one_hot_sentence(tokenized_labeled_sentence["subjects"], character2index),
        "tokens"    : encu.one_hot_sentence(tokenized_labeled_sentence["tokens"], token2index)
    } for tokenized_labeled_sentence in tokenized_labeled_sentences
]
hot_encoded_tokenized_labeled_sentences
```

### Separate into input and label

```python
hot_encoded_tokenized_labels = [hot_encoded_tokenized_labeled_sentence["subjects"] for hot_encoded_tokenized_labeled_sentence in hot_encoded_tokenized_labeled_sentences]
hot_encoded_tokenized_tokens = [hot_encoded_tokenized_labeled_sentence["tokens"] for hot_encoded_tokenized_labeled_sentence in hot_encoded_tokenized_labeled_sentences]
```

```python
X = encu.list2ndarray(hot_encoded_tokenized_labels)
Y = encu.list2ndarray(hot_encoded_tokenized_tokens)
```

```python
jsu.write_json(X.tolist(), "10_X_subjects_black_clover.json")
jsu.write_json(Y.tolist(), "10_Y_tokens_black_clover.json")
```
